```python
# ────────────────────────────────────────────────────────────────
# Cell 0: Fix project paths so src/ and data/ are always found
# Run this FIRST in EVERY notebook
# ────────────────────────────────────────────────────────────────
import sys
import os
from pathlib import Path

# Get the project root (one level up from notebooks/)
notebook_dir = Path(os.getcwd())
project_root = notebook_dir.parent.resolve()  # goes from notebooks/ → root

# Add project root to Python's import path
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Also change current working directory to project root (safer for relative paths)
os.chdir(project_root)

print("Fixed paths:")
print("  • Project root     :", project_root)
print("  • Current dir now  :", os.getcwd())
print("  • sys.path updated → src/ should now be importable")
```

```python
# After the path fix cell
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.data.load_data import load_raw_data

sns.set(style="whitegrid")
%matplotlib inline

df = load_raw_data()
df.head(3)
```

```python
# Churn distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Churn")
plt.title("Churn Distribution (Yes vs No)")
plt.show()
```

```python
# Tenure by churn
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Churn", y="tenure")
plt.title("Tenure Distribution by Churn Status")
plt.show()
```
